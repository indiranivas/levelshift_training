#### 1. Student Marks Analyzer

```python
def avg(marks):
    return sum(marks)//len(marks)
def analyze(marks):

    print("Mark List: ",marks)
    print("Average Marks: " , avg(marks))
    print("Highest Marks: " , max(marks))
    print("lowest Marks: " ,min(marks))
    for i in range(len(marks)):
        if marks[i] < 40 :
            print("Student ",i+1 ,": Fail")
        else :
            print("Student ",i+1 ,": Pass")
marks = []
for i in range(5):
    val  = "Enter marks for student " + str(i+1) + " :"
    mark = int(input(val))
    marks.append(mark)
print("Output :")
print("")
analyze(marks)



```

#### 2. Number Classification Program (*args + Comprehension)

```python
def classify_numbers(*arg):
    even = list(filter(lambda a : a % 2 == 0 ,arg))
    odd  = list(filter(lambda a : a % 2 != 0 ,arg))
    negative  = list(filter(lambda a : a < 0 ,arg))
    print("Even Numbers:",even)
    print("Odd Numbaer:" ,odd)
    print("Negative Number:",negative)
   
print("Output :")
print("")
classify_numbers(10, -5, 7, 8, 3, -2, 12)
```

#### 3. Employee Salary Calculator (kwargs)

```python
def calculate_salary(**kwarg):
    val = kwarg
    print("Employee Name: " , val["name"])
    base = val["base_salary"]
    print("Base Salary:" , base)
    Bonus= val["base_salary"] *  (val["bonus_percent"])//100
    print("Bonus:", Bonus )
    tax = val["base_salary"] *  (val["tax_percent"])//100
    print("Bonus:", tax )
    Final_Salary = base + Bonus - tax
    print("Final Salary :" ,Final_Salary)
print("Output :")
print("")
calculate_salary(name="Arun", base_salary=50000, bonus_percent=10, tax_percent=5)
```

#### 4. Data Cleaning Program (Exception Handling + Comprehension)

```python
def avg(num):
    return sum(num)/len(num)
print("Output :")
print("")
data = ["10", "20", "abc", "40", "xyz", "60"]
lis = []
for i in data:
    try :
        num = int(i)
        lis.append(num)
    except :
        continue
print("Cleaned Data:",lis)
print("")
print("Sum of Numbers:" , sum(lis))
print("Average:",avg(lis))
```

#### 5. Word Frequency Counter (Dictionary + Loop)

```python
def counter(s):
    map = {}
    for ch in s:
        if ch in map:
            map[ch]+=1
        else :
            map[ch] =1
    print("Word Frequency:")
    for key , val in map.items():
        print(key , ":",val)
inp = input().split() #python is easy and python is powerful
print("Output :")
print("")
counter(inp)
```

#### 6. Sorting Students using Lambda

```python
def mark_sort(student):
    return sorted(student , key = lambda a : a[1],reverse=True)
def name_sort(student):
    return sorted(student , key = lambda a : a[0])
val = [('Alice', 92), ('David', 90), ('John', 85), ('Bob', 78)]
print("Output :")
print("")
print("Sorted by Marks (Descending):" )
print(mark_sort(val))
print("")
print("Sorted by Name:")
print(name_sort(val))
```

#### Mini Banking System (Functions + Exception Handling)

```python
def deposit(amount , balance ):
    balance += amount
    return balance
def withdaw(amount,balance):
    if balance - amount >= 0 :
        balance -= amount
        return balance
    else :
        return "Insufficient balance!"

state = 0

balance = 1000
print("Output :")
print("")
while state !=4:

    state = int(input('''
    1. Deposit
    2. Withdraw
    3. Check Balance
    4. Exit
    '''))
    print("Enter choice:",state)
    if state == 1 :
        try :
            amount = int(input("Enter amount :"))
            print("Enter amount: ",amount)
            balance = deposit(amount,balance)
            print("Amount Deposited :" , amount)
            print("Current Balance :" , balance)
        except :
            print("invalid input ")
    elif state ==2 :
        try :
            amount = int(input("Enter amount :"))
            print("Enter amount: ",amount)
            response = withdaw(amount , balance)
            if type(response) is int:
                balance = response
                print("Amount Withdraw :" , amount)
                print("Current Balance :" , balance)
            else :
                print("Insufficient balance!")
        except :
            print("invalid input ")

    elif state ==3 :
        print("Current Balance :" , balance)
    elif  state ==4  :
        print("Exited ")
    else:
        print("invalid input ")
```

#### 9. CSV File Student Result Analyzer

```python
import csv
lis =[]
with open('students.csv', mode='r') as file:
    reader = csv.reader(file)

    for row in reader:
        lis.append(row)

sum = 0
for i in lis[1:]:
    sum += int(i[1])
fail  = lis[:1]  
avg = sum // (len(lis)-1)
print("Output :")
print("")
print(sum)
print("Average Marks:" ,avg)
print("")
print("Passed Students:")
for ch in lis[1:]:
    if int(ch[1]) >= 40:
        print(ch[0])
print("")
print("Failed Students:")

for ch in lis[1:]:
    if int(ch[1]) < 40:
        print(ch[0])
        fail.append(ch)
with open('fail.csv', mode='w', newline='') as file:
    writer = csv.writer(file)
    writer.writerows(fail)
print("")
print("fail.csv is created")
with open('fail.csv', mode="r") as file:
    failed = csv.reader(file)
    for ch in failed:
        print(ch)
```
