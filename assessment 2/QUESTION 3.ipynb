```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
students = [ 
    {"name": "John", "marks": 85, "attendance": 90}, 
    {"name": "Jane", "marks": None, "attendance": 95}, 
    {"name": "Bob", "marks": 110, "attendance": 80}, 
    {"name": "Alice", "marks": 70, "attendance": None}, 
    {"name": "", "marks": 60, "attendance": 85} 
] 
```

```python
pd_students = pd.DataFrame(students)
pd_students
```

```python
pd_students['marks'] = pd_students['marks'].fillna(pd_students['marks'].mean())
pd_students
```

```python
pd_students.info()
```

```python
pd_students = pd_students[pd_students['name'] != '']
```

```python
pd_students['attendance'] = pd_students['attendance'].fillna(pd_students['attendance'].median())
pd_students
```

```python
pd_students = pd_students[pd_students['marks'] <= 100]
pd_students
```

```python
pd_students = pd_students[pd_students['name'] != '']
```

```python
pd_students
```

```python
pd_students['grade'] = pd_students['marks'].apply(lambda x: 'A' if x >= 80 else ('B' if x >= 60 else 'C'))
pd_students
```

```python

pd_students.to_json('Cleaned_Data.json')
```

```python
data = pd.read_json('Cleaned_Data.json')
print(data)
```

```python
avg_marks = pd_students['marks'].mean()
toper = pd_students[pd_students['marks'] == pd_students['marks'].max()]['name'].values[0]
print(f"Average: {avg_marks}")
print(f"Topper: {toper}")
```

```python
print("Grade Distribution:")
grade_counts = pd_students['grade'].value_counts()
print(grade_counts)
```

```python
print("Cleaned Data:\n",data)

print("\nAverage Marks:", avg_marks)

print("\nTopper:", toper)

print("\nGrade Distribution:")
for grade, count in grade_counts.items():
    print(f"{grade}: {count}")
```
