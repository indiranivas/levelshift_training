```python
df = df.dropna()
df = df[df['response_time'] !='abc ']
df['response_time'] = df['response_time'].astype(int)
df
```

```python
# remove 
log_levels = df['log_level'].value_counts()
print(log_levels)
```

```python
def catagory(df):
    if df["response_time"] < 150:
        return "FAST"
    elif df["response_time"] <= 300:
        return "MEDIUM"    
    else:
        return "SLOW"
df["category"] = df.apply(catagory, axis=1)
df
```

```python
average_response_time = df['response_time'].mean()
no_slow_responses = df[df['category'] == 'SLOW'].shape[0]
```

```python
'''ERROR LOGS: 
10:02 → Invalid response_time 
10:04 → Missing response_time'''
def erlog(df1):
    if df1["response_time"] == 'abc ':
        return "Invalid response_time"
    elif pd.isna(df1["response_time"]):
        return "Missing response_time"
    else:
        return "No error"
```

```python
df1
df1["error_log"] = df1.apply(erlog, axis=1)
df1
```

```python
print("LOG LEVEL COUNT: ")
for log_level, count in log_levels.items():
    print(f"{log_level}: {count}")
print()
print("AVERAGE RESPONSE TIME: " , average_response_time)
print()
print("SLOW REQUEST:", no_slow_responses)
print()
print("ERROR LOGS: ")
for index, row in df1.iterrows():
    if row["error_log"] != "No error":
        print(f"{row['timestamp']} → {row['error_log']}")
```
