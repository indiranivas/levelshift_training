#   5. PCA (Dimensionality Reduction) 


```python
import pandas as pd
import numpy as np 
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
df= pd.read_csv('cust.csv')
df
```

```python
scaler = StandardScaler()
```

```python
x = df[['Age','Income','Spending','Savings','Transactions']]
scaled_x = scaler.fit_transform(x)

pca =PCA(n_components=2)

d2 =pca.fit_transform(scaled_x)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure()
plt.scatter(d2[:, 0], d2[:, 1])
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title("PCA Visualization")
plt.show()
```

# 6. PCA (Variance Explained Concept) 

```python
pca1 =PCA()

d3 =pca1.fit_transform(scaled_x)
```

```python
explained_var = pca.explained_variance_ratio_

print(sum(pca1.explained_variance_ratio_))
```

```python
cumulative_var = np.cumsum(explained_var)
print("\nCumulative Variance:\n", cumulative_var)

n_components_95 = np.argmax(cumulative_var >= 0.95) +1
print("\nNumber of components for 95% variance:", n_components_95)
```

```python
import matplotlib.pyplot as plt

plt.plot(cumulative_var, marker='o')
plt.axhline(y=0.95, linestyle='--')  
plt.xlabel("Number of Components")
plt.ylabel("Cumulative Explained Variance")
plt.title("PCA Variance Explained")
plt.show()
```
