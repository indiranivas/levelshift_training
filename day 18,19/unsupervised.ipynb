# 1. K-Means Clustering (Customer Segmentation) 


```python
import pandas as pd 
import numpy as np 
from sklearn.cluster import KMeans
```

```python
df = pd.read_csv('customer_dataset.csv')
data = df[['Annual Income','Spending Score']]
```

```python
df.info()
```

```python
import matplotlib.pyplot as plt
```

```python
n= 2
while n <7:
    kmeans = KMeans(n_clusters=n, random_state=0)
    kmeans.fit(data)

    # Output cluster centers and labels
    print("Cluster Centers:\n", kmeans.cluster_centers_)
    print("Labels:", kmeans.labels_)

    # Visualize the clusters
    plt.figure()
    plt.scatter(data.iloc[:, 0], data.iloc[:, 1], c=kmeans.labels_, cmap='viridis')
    plt.scatter(kmeans.cluster_centers_[:, 0], 
                kmeans.cluster_centers_[:, 1], 
                s=200, marker='X')
    
    


    plt.xlabel("Annual Income")
    plt.ylabel("Spending Score")
    plt.title(f"K-Means Clustering (k={n})")
    plt.show()
    n+=1
```

```python
def label_cluster(row):
    if row["Cluster"] == 0:
        return "High Income - Low Spending"
    elif row["Cluster"] == 1:
        return "Low Income - Low Spending"
    else:
        return "Low Income - High Spending"
```

```python
kmeans1 = KMeans(n_clusters=3, random_state=0)
df["Cluster"] = kmeans1.fit_predict(data)


df["Segment"] = df.apply(label_cluster, axis=1)
cluster_summary = df.groupby("Segment")[["Age", "Annual Income", "Spending Score"]].mean()


print("\nCluster Summary:\n", cluster_summary)
df.head()
```

# 2. Elbow Method (Find Optimal K) 

```python

from sklearn import metrics
from scipy.spatial.distance import cdist
```

```python
distortions = []
inertias = []
mapping1 = {}
mapping2 = {}
K = range(1, 11)

for k in K:
    kmeanModel = KMeans(n_clusters=k, random_state=42).fit(data)
    
    distortions.append(sum(np.min(cdist(data, kmeanModel.cluster_centers_, 'euclidean'), axis=1)**2) / data.shape[0])
    
    inertias.append(kmeanModel.inertia_)
    
   
    mapping2[k] = inertias[-1]
```

```python
print("Inertia values:")
for key, val in mapping2.items():
    print(f'{key} : {val}')

plt.plot(K, inertias, 'bx-')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia')
plt.title('The Elbow Method using Inertia')
plt.show()
```

#  3. Silhouette Score (Cluster Evaluation) 

```python
from sklearn.metrics import silhouette_score
```

```python
map_score ={}
n=2
score = []
while n < 7:    
    kmean_ss = KMeans(n_clusters=n)
    predicted_labels = kmean_ss.fit_predict(data)
    silhouette_val = silhouette_score(data, predicted_labels)
    map_score[n] = silhouette_val
    print( n,round((silhouette_val),2))
    score.append(silhouette_val)
    n+=1

    
```

```python
for key, val in map_score.items():
    print(f'{key} : {val}')

x = list(map_score.keys())
y= list(map_score.values())

fig = plt.figure()
ax = fig.add_subplot(111)

x = list(map_score.keys())
y= list(map_score.values())
line, = ax.plot(x, y)

ymax = max(y)
xpos = y.index(ymax)
xmax = x[xpos]

ax.annotate(f'best K : {xmax}', xy=(xmax, ymax), xytext=(xmax, ymax+0.02),
            arrowprops=dict(facecolor='black', shrink=0.00),
            )

ax.set_ylim(0.3,0.45)
plt.show()
```

# 4. Hierarchical Clustering (Concept + Dendrogram) 

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import AgglomerativeClustering
from sklearn.metrics import silhouette_score
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
linked = linkage(df, method='ward')

plt.figure(figsize=(12, 6))
dendrogram(
    linked,
    truncate_mode='lastp',
    p=30
)
plt.title("Hierarchical Clustering Dendrogram")
plt.xlabel("Cluster Size")
plt.ylabel("Distance")
plt.show()
```

```python
hc_model = AgglomerativeClustering(
    n_clusters=10,
    linkage='ward'
)
```

```python
cluster_labels = hc_model.fit_predict(df)
```

```python
pd.Series(cluster_labels).value_counts()
```

```python
sil_score = silhouette_score(df, cluster_labels)
print("Score :", round(sil_score, 2))
```
