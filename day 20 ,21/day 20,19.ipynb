Q1: Save & Load Model (Pickle vs Joblib)


```python
import pickle
import joblib
from sklearn.linear_model import LinearRegression

X = [[1000, 2], [1500, 3]]
y = [200000, 300000]

model = LinearRegression()
model.fit(X, y)

with open('model.pkl', 'wb') as f:
    pickle.dump(model, f)

joblib.dump(model, 'model.joblib')

with open('model.pkl', 'rb') as f:
    loaded_model_pickle = pickle.load(f)

loaded_model_joblib = joblib.load('model.joblib')

new_data = [[1200, 3]]
prediction = loaded_model_joblib.predict(new_data)[0]
print(f"Prediction: {int(prediction)}")
```

Q2: Build Flask Prediction API


```python
from flask import Flask, request, jsonify
import joblib

app = Flask(__name__)
model = joblib.load('model.joblib')

@app.route('/predict', methods=['POST'])
def predict():
    data = request.get_json()
    features = data['features']
    prediction = model.predict([features])[0]
    return jsonify({'prediction': int(prediction)})

if __name__ == '__main__':
    app.run(debug=True)
```

Q3: API Error Handling


```python
from flask import Flask, request, jsonify
import joblib

app = Flask(__name__)
model = joblib.load('model.joblib')

@app.route('/predict', methods=['POST'])
def predict():
    try:
        data = request.get_json()
        features = data['features']
        
        if not isinstance(features, list) or len(features) != 2:
            raise ValueError
        if not all(isinstance(x, (int, float)) for x in features):
            raise ValueError
            
        prediction = model.predict([features])[0]
        return jsonify({'prediction': int(prediction)})
    
    except Exception as e:
        return jsonify({'error': 'Invalid input format'}), 400

if __name__ == '__main__':
    app.run(debug=True)
```

```python
import joblib
from datetime import datetime
from sklearn.linear_model import LinearRegression

X = [[1000, 2], [1500, 3]]
y = [200000, 300000]

model = LinearRegression()
model.fit(X, y)

version = datetime.now().strftime("%Y%m%d")
versioned_filename = f'model_v{version}.pkl'

joblib.dump(model, versioned_filename)
joblib.dump(model, 'model_latest.pkl')

print(f"{versioned_filename} created")
print("model_latest.pkl created")
```

Q6: Build Prediction UI


```python
import streamlit as st
import joblib
import numpy as np

model = joblib.load('model_latest.pkl')

st.title("House Price Predictor")

area = st.number_input("Area (sq ft)", min_value=0)
bedrooms = st.number_input("Bedrooms", min_value=0, step=1)

if st.button("Predict"):
    prediction = model.predict([[area, bedrooms]])[0]
    st.write(f"Predicted Price: {int(prediction)}")
```
