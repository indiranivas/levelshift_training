Q12: Stemming vs Lemmatization


```python
from nltk.stem import PorterStemmer, WordNetLemmatizer
import nltk
nltk.download('wordnet')
nltk.download('omw-1.4')

text = "running studies better"
words = text.split()

stemmer = PorterStemmer()
stemmed = [stemmer.stem(word) for word in words]

lemmatizer = WordNetLemmatizer()
lemmatized = [lemmatizer.lemmatize(word, pos='v') for word in words]

print(f"Stemming: {stemmed}")
print(f"Lemmatization: {lemmatized}")
```

### ECTION 4: NLP FEATURE ENGINEERING (Q15–Q18)


Q15: Bag of Words


```python
from sklearn.feature_extraction.text import CountVectorizer

documents = ["I love AI", "AI is great"]

vectorizer = CountVectorizer()
X = vectorizer.fit_transform(documents)

vocabulary = vectorizer.get_feature_names_out()
matrix = X.toarray()

print(f"Vocabulary: {list(vocabulary)}")
print(f"Matrix: {matrix.tolist()}")
```

Q16: TF-IDF


```python
from sklearn.feature_extraction.text import TfidfVectorizer

documents = ["I love AI", "AI is great", "AI is awesome"]

vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(documents)

feature_names = vectorizer.get_feature_names_out()
matrix = X.toarray()

print("TF-IDF Matrix:")
for i, doc in enumerate(matrix):
    print(f"Document {i+1}: {dict(zip(feature_names, doc))}")

print("\nAI gets lower weight (appears frequently)")
```

Q17: Build Sentiment Model


```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
import numpy as np

texts = ["This is amazing", "This is terrible", "Great product", "Bad experience", "I love it", "I hate it"]
labels = [1, 0, 1, 0, 1, 0]

vectorizer = CountVectorizer()
X = vectorizer.fit_transform(texts)

model = MultinomialNB()
model.fit(X, labels)

new_text = ["This is amazing"]
new_text_vectorized = vectorizer.transform(new_text)
prediction = model.predict(new_text_vectorized)[0]

result = "Positive" if prediction == 1 else "Negative"
print(result)
```

Q18: Pipeline Creation




```python
from sklearn.pipeline import Pipeline
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
import re
from sklearn.base import BaseEstimator, TransformerMixin

class TextCleaner(BaseEstimator, TransformerMixin):
    def fit(self, X, y=None):
        return self
    
    def transform(self, X, y=None):
        cleaned_texts = []
        for text in X:
            text = text.lower()
            text = re.sub(r'[^a-zA-Z\s]', '', text)
            cleaned_texts.append(text)
        return cleaned_texts

texts = ["I LOVE this!!!", "This is terrible :(", "Amazing product!", "Bad experience"]
labels = [1, 0, 1, 0]

pipeline = Pipeline([
    ('cleaner', TextCleaner()),
    ('vectorizer', TfidfVectorizer()),
    ('classifier', MultinomialNB())
])

pipeline.fit(texts, labels)

new_text = ["This is amazing"]
prediction = pipeline.predict(new_text)[0]

result = "Positive" if prediction == 1 else "Negative"
print(f"Single pipeline working end-to-end: {result}")
```
