# Question 2: Feature Engineering (Polynomial + Interaction Features)

## 1. Why do Polynomial Features Help?

### 🔹 Concept

Polynomial features allow models to capture **non-linear relationships** between variables.

👉 Many real-world problems are **not linear**  
Example:
- Salary vs Churn is not a straight line  
- Interaction matters: Salary × Tenure  

---

### 🔹 How it works

Original features:
\[
x_1, x_2
\]

After polynomial transformation (degree = 2):
\[
x_1, x_2, x_1^2, x_2^2, x_1 x_2
\]

---

### 🔹 Why useful?

- Captures **feature interactions**
- Models **curved relationships**
- Improves performance of linear models

---

### 🔥 Example

Instead of:
- Salary alone ❌  

We get:
- Salary²  
- Salary × Tenure ✅  

👉 Much richer information

---

## 2. What is Overfitting Risk with Too Many Features?

### 🔹 Concept

Overfitting happens when:
👉 Model learns **training data too well (including noise)**

---

### 🔹 Why too many features cause it?

Polynomial features can explode:

\[
n \rightarrow \frac{(n + d)!}{n! \cdot d!}
\]

Where:
- \( n \) = original features  
- \( d \) = degree  

👉 Feature count grows very fast

---

### 🔹 Problem

- Model memorizes instead of learning  
- Performs well on training ❌  
- Performs poorly on test data ❌  

---

### 🔹 Symptoms

- High training accuracy  
- Low test accuracy  

---

### 🔥 Solution

- Feature selection  
- Regularization  
- Limit polynomial degree  

---

## 3. What are Domain-Based Features?

### 🔹 Concept

Domain-based features are:
👉 Features created using **real-world knowledge of the problem**

---

### 🔹 Examples (Bank Churn)

Instead of raw data:
- Salary  
- Balance  

We create:
- **Balance / Salary ratio**
- **Tenure group (new vs old customer)**
- **Active vs inactive flag**

---

### 🔹 Why important?

- Adds **meaningful information**
- Improves model accuracy
- Reduces need for complex models

---

### 🔥 Real Insight

> Domain knowledge often beats complex algorithms

---

## ✅ Summary

| Concept | Key Idea |
|--------|--------|
| Polynomial Features | Capture non-linearity |
| Overfitting | Too many features = poor generalization |
| Domain Features | Use real-world knowledge |

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel('house_price.xlsx')
df
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly = PolynomialFeatures(degree=2, include_bias=False)
poly_features = poly.fit_transform(df)
feature_names = poly.get_feature_names_out(df.columns)

df_poly = pd.DataFrame(poly_features, columns=feature_names)
df_poly = df_poly[['Area' ,'Bedrooms' , 'Price' ,'Area^2' ,'Bedrooms^2','Area Bedrooms']]
print("\nPolynomial Features DataFrame:")
print(df_poly)
df_poly = df_poly.astype('int32')
```
