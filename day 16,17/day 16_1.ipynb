# Question 1: Feature Selection using Correlation & Chi-Square

## 1. Difference between Correlation vs Chi-Square

### 🔹 Correlation
- Measures **linear relationship** between two variables
- Works on **continuous numerical data**
- Output range: **-1 to +1**

#### Interpretation:
- +1 → Strong positive relationship  
- -1 → Strong negative relationship  
- 0 → No relationship  

#### Example:
- Salary ↑ → Churn ↓ → Negative correlation  

---

### 🔹 Chi-Square Test
- Measures **dependency between variables**
- Works on **categorical data**
- Output: **Chi-square score (≥ 0)**

#### Interpretation:
- High score → Strong dependency  
- Low score → Weak dependency  

---

### 🔥 Key Differences

| Feature        | Correlation                  | Chi-Square                  |
|---------------|-----------------------------|-----------------------------|
| Data Type     | Continuous                  | Categorical                 |
| Relationship  | Linear                      | Dependency (non-linear too) |
| Output        | -1 to +1                    | 0 to ∞                      |
| Use Case      | Numeric features            | Categorical features        |

---

## 2. When to Use Chi-Square

### ✅ Use Chi-Square when:
- Target variable is **categorical** (e.g., Churn: 0/1)
- Features are:
  - Categorical OR
  - Converted to non-negative values (scaled)

### 🔹 Example Use Cases:
- Gender vs Churn  
- Product Type vs Purchase  
- Region vs Sales Category  

---

### ⚠️ Important Conditions:
- Data must be **non-negative**
- Works best with **discrete/categorical features**
- For continuous data → apply:
  - **Binning** OR
  - **MinMax Scaling**

---

## 3. Why Remove Low-Impact Features?

### 🔹 Definition:
Low-impact features = features with:
- Low correlation  
- Low Chi-square score  

---

### 🚫 Problems if NOT removed:

1. **Noise in model**
   - Confuses learning process  

2. **Overfitting**
   - Model learns useless patterns  

3. **Slower training**
   - More features = more computation  

4. **Reduced accuracy**
   - Irrelevant data affects predictions  

---

### ✅ Benefits of Removing Them:

- Better model performance 🚀  
- Faster training ⏱️  
- Improved generalization  
- Simpler and interpretable model  

---

### 🔥 Key Insight

> “Good features = Better model than complex algorithms”

---

## ✅ Summary

- **Correlation** → Linear relationship (numerical data)  
- **Chi-Square** → Dependency (categorical data)  
- **Feature Selection** → Keep useful, remove noise  

---

```python
import pandas as pd 
import  numpy as np 
```

```python
df =  pd.read_excel("churn.xlsx")
df
```

```python
correlation_matrix = df.corr()
correlation_matrix
```

```python
churn_corr = correlation_matrix['Churn']
```

```python
selected_features = churn_corr[abs(churn_corr) > 0.5]
selected_features = selected_features.drop("Churn")
print(selected_features)
```

```python
from sklearn.feature_selection import chi2
from sklearn.preprocessing import MinMaxScaler

X = df.drop("Churn", axis=1)
y = df["Churn"]

scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)

chi_scores, p_values = chi2(X_scaled, y)

chi_df = pd.DataFrame({
    "Feature": X.columns,
    "Score": chi_scores
})

chi_df = chi_df.sort_values(by="Score", ascending=False)

print(chi_df)
```

```python
top_features = chi_df.head(2)
print("\nTop 2 Features:\n", top_features)
```
