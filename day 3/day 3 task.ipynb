```python
class Course:
    def __init__(self, course_id , course_name , max_students=3 ):
        self.course_id = course_id
        self.course_name = course_name
        self.instructor = None
        self.max_students = max_students
        self.students_enrolled = []

    def add_student(self , student):
        if student in self.students_enrolled:
            print(f"{student.name} already enrolled in {self.course_name}")
            return

        if len(self.students_enrolled) >= self.max_students:
            print("Course is full")
            return

        self.students_enrolled.append(student)
        print(f"{student.name} successfully enrolled in {self.course_name}")

    def course_details(self):
        print("----- Course Details -----")
        print("Course ID:", self.course_id)
        print("Course Name:", self.course_name)
        print("Instructor:", self.instructor.name if self.instructor else "Not Assigned")
        print("Students Enrolled:", len(self.students_enrolled))


class Person:
    def __init__(self,name , email):
        self.name = name
        self.email = email
    
    def display_info(self):
        print("Name:", self.name)
        print("Email:", self.email)


class Instructor(Person):
    def __init__(self, name, email ,specilization):
        super().__init__(name, email)
        self.specilization = specilization

    def assign_course(self, course):
        course.instructor = self
        print(f"Instructor {self.name} assigned to {course.course_name}")


class Student(Person):
    def __init__(self, name, email , student_id):
        super().__init__(name, email)
        self.student_id = student_id
        self.courses = []

    def enroll(self, course):
        course.add_student(self)
        if course not in self.courses and len(course.students_enrolled) <= course.max_students:
            self.courses.append(course)

    def view_courses(self):
        print(f"----- {self.name}'s Courses -----")
        for c in self.courses:
            print(c.course_name)


students = []

def inst():
    course_id = input("Enter course id: ")
    course_name = input("Enter course name: ")

    course = Course(course_id, course_name)

    inst_name = input("Enter instructor name: ")
    inst_email = input("Enter instructor email: ")
    specialization = input("Enter specialization: ")

    instructor = Instructor(inst_name, inst_email, specialization)
    instructor.assign_course(course)

    return course


def stu(course):
    name = input("Enter student name: ")
    email = input("Enter student email: ")
    sid = input("Enter student id: ")

    student = Student(name, email, sid)
    students.append(student)

    student.enroll(course)


python = inst()

stu(python )
stu(python )
stu(python )  
stu(python )

python.course_details()

students[0].view_courses()


```
